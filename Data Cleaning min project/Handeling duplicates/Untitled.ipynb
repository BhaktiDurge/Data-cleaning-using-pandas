```python
# handeling duplicate data 
# Duplicate Data : entier row is repeated in data called duplicated data
```

```python
import pandas as pd
```

```python
data = { "name": ['a','b','c','d','a','c'] , "English" : [8,7,5,8,8,5], "Math" : [2,3,4,5,2,6]}
df = pd.DataFrame(data)
df
```

```python
# As we can see row 4th is repeated
# but to check for large data let's make new colunm duplicated
# duplicated() gives bool values as output

df.duplicated()
df["duplicated"] = df.duplicated()
df
```

```python
# To remove duplicates use drop_duplicates fun
# If duplicates of same row is mulltiple then use keep attribute keep = True
# And use inplace tto make change is orignal df
# first remove colunm duplicated caues it will be also compared while checking duplicated
 
df.drop("duplicated", axis=1, inplace=True)
df.drop_duplicates(inplace=True)
df
```

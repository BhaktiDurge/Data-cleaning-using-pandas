```python
# Filling missing data in the dataset
# We use the fillna() function to handle missing (NaN) values.
# There are three common methods we will use:
#
# 1) Forward Fill (ffill):
#    - This method fills missing values using the previous valid value
#    - It carries the last observed value forward
#
# 2) Backward Fill (bfill):
#    - This method fills missing values using the next valid value
#    - It takes the next observed value and fills backward
#
# 3) Mode:
#    - This method fills missing values using the most frequently
#      occurring value (mode) in the column
#
# We will apply each method step by step using the fillna() function.
```

```python
# 1) Forward Fill (ffill) just not using inplace attribute now cause don't want change in df1
```

```python
df1.ffill()
```

```python
# 2) Backward Fill (bfill): and just not using inplace attribute now cause don't want change in df1
```

```python
df1.bfill()
```

```python
# 3) Mode:
```

```python
# let's take city colunm for this
# here df1["city"].mode() gives frequent value in city colunm that value is fill in city colunm where there is missing value 
# just not using inplace attribute now cause don't want change in df1 in fillna fun
df1["City"].fillna(df1["City"].mode()[0])
```

```python
# To fill missing values with mode for all categorical/object columns using a loop

for i in df1.select_dtypes(include="object").columns:
    df1[i].fillna(df1[i].mode()[0], inplace=True)
```

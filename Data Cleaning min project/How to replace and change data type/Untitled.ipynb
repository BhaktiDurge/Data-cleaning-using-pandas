```python
import pandas as pd
```

```python
# Handeling mixed data like colunm is of age and it has value like 2five 
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# Now form above we get to know that we have null value 
# So 1st filling null values using mode method 

df["rating"].fillna(df["rating"].mode()[0], inplace=True)
```

```python
# let's check now again for null values

df.isnull().sum()
```

```python
# now let's check which values are repeating in rating colunm

df["rating"].value_counts()
```

```python
# NOW we will replace this mixed data using replace fun 
# NOTE : WE HAVE NOT CHANGE THE DATA TYTPE OF OF RATING COULNM YET IT'S STILL OBJECT

df["rating"].replace("3.5stars","3.5", inplace=True)
df["rating"].replace("Five","5.0", inplace=True)
df["rating"].replace("5five","5.5", inplace=True)
df["rating"].replace("four","4.0", inplace=True)
```

```python
# again check for
df["rating"].value_counts()
```

```python
#Now let's convert datatype of rating colunm using astype
df["rating"].astype("float64")
df["rating"].dtype
```
